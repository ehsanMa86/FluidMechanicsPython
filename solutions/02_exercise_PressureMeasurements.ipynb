The closed tank contains a liquid and its vapour. The liquid density is $800 \,\, \mathrm{kg/m^3}$ and its vapour density is negligible. The pressure at the top of the liquid in the tank is $120$ kPa abs and atmospheric pressure is 101 kPa abs. If the specific gravity of mercury is 13.6 and the density of water is $1000 \,\, \mathrm{kg/m^3}$, what is the reading of h on the mercury manometer.

```python
Image("./../resources/tank.png")
```

### Solution:
***

$$p_v = \gamma_l (1) + \gamma_{Hg} h + p_{atm}$$

$$ h = \frac{p_v + \gamma_l (1) - p_{atm}}{\gamma_{Hg}} = $$

$$ h = \frac{p_v + \rho_l g (1) - p_{atm}}{SG_{Hg} \rho g} = \frac{120 \times 10^3 + 800 \times 9.81 \times 1 - 101 \times 10^3}{13.6 \times 1000 \times 9.81}$$

$$h = 0.201 m = 201 mm$$
