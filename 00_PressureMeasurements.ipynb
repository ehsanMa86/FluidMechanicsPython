# Pressure Measurements

Pressure is a normal force per unit area
$$ P = \frac{F}{A} $$

**_unit_**:

**SI**: $N/m^2$ (Pa)

**English Engineering unit**: $lbf/in^2$ (psi)

<center>6895 Pa = 1 psi</center>

Because the Pa is an small number, we usually use kPa unit which kPa = 1000 Pa .

There are three ways that pressure can be reported:
* absolute pressure
* gage pressure
* vacuum pressure

### Absolute Pressure:

Absolute pressure is measured relative to a perfect vacuum. Zero absoulute pressure corresponds to a perfect vacuum.

**Units:** psia or kPa abs

### Gage Pressure:

Gage pressure is measured relative to the local atmospheric pressure. It is related to absolute pressure by

$$P_{gage} = P_{absolute} - P_{atm}$$

So a gage pressure of zero means the pressure is the local atmospheric pressure.

**Units:** psig, psi, kPa

### Vacuum Pressure:

If gage pressure is negative (i.e. below atmospheric) expresssed as a positive number. Vacuum pressure is also measured relative to the local atmospheric pressure, but it is reported as how far below local atmospheric pressure something is. That is

$$P_{vacuum} = P_{atm} - P_{absolute}$$

**Units:** psi vac or kPa vac

$$ \rho_{Hg} = 13.6 \times 1000 = 13,600 \,\,\mathrm{kg/m^3}$$

# Manometry:
***

Recall that for an incompressible fluid, pressure is dependant on fluid depth or height. This characteristic can be used to measure pressure.

A **_manometer_** is a vertical or inclined tube containing a fluid, and is one type of pressure measurment device.

### Example:
***

Let's consider a simple case here. So here we have an open column which is filled with gasoline at the top and it's overlaying an immiscible fluid, water, so the two fluids don't mix. What is the pressure at the bottom of these two columns?

```python
Image("resources/differentDensity.png")
```

Gasoline: $p_1 - p_0 = \gamma_g (z_0 - z_1)$

Water   : $p_2 - p_1 = \gamma_w (z_1 - z_2)$

Add     : $p_2 - p_0 = \gamma_g (z_0 - z_1) + \gamma_w (z_1 - z_2)$

Or

$p_2 = p_0 + \gamma_g (z_0 - z_1) + \gamma_w (z_1 - z_2)$

### U-tube manometer:
***

```python
Image("resources/manometerU.png")
```

We wish to measure the pressure at location $A$. The manometer is filled with fluid, and as $P_A$ varies, the height of the fluid column will vary. By measuring this height, the pressure can be determined.

In general we will follow certain steps to find the pressure at point $A$, i.e. $p_A$
* Start at one end and write pressure there.
* Work to other end, equating to pressures due to each fluid column:
* (+) pressure if next meniscus higher 
* (-) pressure if next meniscus lower 
* (+) pressure at other end

$$p_A = - \gamma_1 h_1 + \gamma_2 h_2 + p_{atm}$$
